```python
def display_two_masks(mask_path_1, mask_path_2):
    mask_1 = np.load(mask_path_1) + 1
    mask_2 = np.load(mask_path_2) + 1
    # display the two masks in the two panels of the same plot
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
    ax1.axis('off')  
    ax2.axis('off')  
    ax1.imshow(mask_1)
    ax2.imshow(mask_2)
    plt.show()

def display_two_masks_with_ids(mask_path_1, id_1, mask_path_2, id_2):
    mask_1 = np.load(mask_path_1)
    mask_1 = np.where(mask_1 == id_1, 1, 0)
    mask_2 = np.load(mask_path_2)
    mask_2 = np.where(mask_2 == id_2, 1, 0)
    # display the two masks in the two panels of the same plot
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
    ax1.axis('off')  
    ax2.axis('off')  
    ax1.imshow(mask_1)
    ax2.imshow(mask_2)
    plt.show()
    

def display_two_image(image_path_1, image_path_2):
    image_1 = plt.imread(image_path_1)
    image_2 = plt.imread(image_path_2)
    # display the two images in the two panels of the same plot
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
    ax1.axis('off')  
    ax2.axis('off')  
    ax1.imshow(image_1)
    ax2.imshow(image_2)
    plt.show()

image_id_1 = 'DJI_0263'
masks_path_1 = "../../data/mission_2_segmentations/{}.npy".format(image_id_1)
image_path_1 = "../../data/mission_2/{}.JPG".format(image_id_1)
overlap_path_1 = "../../data/mission_2_segmentations/{}_overlap.png".format(image_id_1)

image_id_2 = 'DJI_0242'
masks_path_2 = "../../data/mission_2_segmentations/{}.npy".format(image_id_2)
image_path_2 = "../../data/mission_2/{}.JPG".format(image_id_2)
overlap_path_2 = "../../data/mission_2_segmentations/{}_overlap.png".format(image_id_2)


id_1, id_2 = 10, 3
#display_two_masks(masks_path_1, masks_path_2)
display_two_image(image_path_1, image_path_2)
display_two_masks_with_ids(masks_path_1, id_1, masks_path_2, id_2)
```

```python


def display_mask_with_id(masks, id):
    mask = np.where(masks == id, 1, 0)
    plt.imshow(mask)
    plt.show()
    return mask

def erode_mask_with_id(masks, id):
    mask = np.where(masks == id, 1, 0)
    mask = ndimage.binary_erosion(mask, structure=np.ones((5,5)), iterations=2).astype(mask.dtype)
    plt.imshow(mask)
    plt.show()
    return mask

def display_mask_diff(mask1, mask2):
    mask_diff = np.where(mask1 == mask2, 0, 1)
    plt.imshow(mask_diff)
    plt.show()

id = 1
mask1 = display_mask_with_id(masks, id)
mask2 = erode_mask_with_id(masks, id)
display_mask_diff(mask1, mask2)
```
